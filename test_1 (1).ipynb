```python
from openpyxl import load_workbook
import matplotlib.pyplot as plt
import numpy as np
```

```python
def write_xlsx_file(name,sheet):
    data=[]
    wb =  load_workbook(name, read_only=True)
    ws = wb[sheet]
    for row in ws.rows:
        for cell in row:
            if isinstance(cell.value,float):
                data.append(cell.value)    
    for i in range(4):
    x=np.arange(0,len(data[:,i]),1)
    plt.subplot(2,2,i+1)
    plt.plot(x,data[:,i])
    return data
   
```

```python
data=np.reshape(write_xlsx_file('data.xlsx','3'),(-1,4))
for i in range(4):
    x=np.arange(0,len(data[:,i]),1)
    plt.subplot(2,2,i+1)
    plt.plot(x,data[:,i])
```
